```python
from IPython.display import HTML

HTML("""
<!DOCTYPE html>
<html>
<head>
<title>Privacy Risk Analyzer</title>

<style>
body {
    font-family: Arial;
    background: linear-gradient(135deg, #1e3c72, #2a5298);
    color: white;
    text-align: center;
    padding: 30px;
}

.container {
    max-width: 600px;
    margin: auto;
    background: rgba(255,255,255,0.1);
    padding: 20px;
    border-radius: 15px;
    backdrop-filter: blur(10px);
}

h1 {
    margin-bottom: 20px;
}

.card {
    background: rgba(0,0,0,0.3);
    margin: 10px 0;
    padding: 15px;
    border-radius: 10px;
    text-align: left;
}

button {
    padding: 12px 25px;
    border: none;
    background: #00ffcc;
    color: black;
    font-weight: bold;
    border-radius: 8px;
    cursor: pointer;
}

button:hover {
    background: #00ccaa;
}

.progress {
    width: 100%;
    background: #333;
    border-radius: 10px;
    margin-top: 20px;
}

.progress-bar {
    height: 20px;
    width: 0%;
    background: lime;
    border-radius: 10px;
    transition: 0.5s;
}

.result {
    margin-top: 15px;
    font-size: 18px;
    font-weight: bold;
}
</style>
</head>

<body>

<div class="container">
    <h1> Privacy Risk Analyzer</h1>

    <div class="card"> Cookies Status: <span id="cookies"></span></div>
    <div class="card"> Connection Security: <span id="https"></span></div>
    <div class="card"> Internet Status: <span id="online"></span></div>
    <div class="card"> Device Info: <span id="device"></span></div>

    <button onclick="analyze()">Analyze Privacy</button>

    <div class="progress">
        <div class="progress-bar" id="bar"></div>
    </div>

    <div class="result" id="result"></div>
</div>

<script>
function analyze() {
    let score = 100;

    // Cookies
    if (navigator.cookieEnabled) {
        document.getElementById("cookies").innerText = "Enabled ❌ (Tracking risk)";
        score -= 20;
    } else {
        document.getElementById("cookies").innerText = "Disabled ";
    }

    // HTTPS
    if (location.protocol === "https:") {
        document.getElementById("https").innerText = "Secure (HTTPS) ";
    } else {
        document.getElementById("https").innerText = "Not Secure (HTTP) ";
        score -= 30;
    }

    // Online
    if (navigator.onLine) {
        document.getElementById("online").innerText = "Online ";
    } else {
        document.getElementById("online").innerText = "Offline ";
        score -= 10;
    }

    // Device Info
    document.getElementById("device").innerText = navigator.userAgent;

    // Update progress bar
    document.getElementById("bar").style.width = score + "%";

    // Result
    let msg = "";
    if (score >= 80) msg = "🟢 Excellent Privacy";
    else if (score >= 50) msg = "🟡 Moderate Risk";
    else msg = "🔴 High Risk";

    document.getElementById("result").innerText =
        "Privacy Score: " + score + "% → " + msg;
}
</script>

</body>
</html>
""")
```
